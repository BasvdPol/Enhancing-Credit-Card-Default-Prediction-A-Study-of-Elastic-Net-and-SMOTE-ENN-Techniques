```python
cv_outer_f1_score_without = [0.36302142051860203, 0.34595816845675526, 0.3596392333709132, 0.362400906002265, 0.36610169491525424]
```

```python
cv_outer_f1_score_K = [0.8710785943908658, 0.8681306431835916, 0.8662958843159065, 0.8631229235880399, 0.8716186252771618]
```

```python
cv_outer_f1_score_ENN = [0.7801030264005151, 0.7752790810548454, 0.7866921362623368, 0.7843883864826273, 0.7768911586338972]
```

```python
from scipy import stats
# Perform one-way ANOVA
statistic, p_value = stats.f_oneway(cv_outer_f1_score_ENN, cv_outer_f1_score_K, cv_outer_f1_score_without)

print("One-way ANOVA p-value:", p_value)
```

```python
import matplotlib.pyplot as plt
labels = ['ENN', 'K', 'Without Resampling']

# Box plot
plt.figure(figsize=(10, 6))
plt.boxplot([cv_outer_f1_score_ENN, cv_outer_f1_score_K, cv_outer_f1_score_without], labels=labels)
plt.title('F1 Score Performance of the best Elastic Net models')
plt.ylabel('F1 Score')
plt.xlabel('Model')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
labels = ['ENN', 'K']

# Box plot
plt.figure(figsize=(10, 6))
plt.boxplot([cv_outer_f1_score_ENN, cv_outer_f1_score_K], labels=labels)
plt.title('F1 Score Performance of the best resampled Elastic Net models')
plt.ylabel('F1 Score')
plt.xlabel('Model')
plt.grid(True)
plt.show()
```
